```python
import pandas as pd
```

```python
df = pd.read_excel('March_Sept_2022_admission_times.xlsx')
df
```

```python
ser = pd.to_datetime(df['estimated_admission_dttm'], utc=True)
ser = ser.dt.tz_convert('US/Pacific')
df['estimated_admission_dttm'] = ser
df['estimated_admission_dttm']
```

```python
ser = pd.to_datetime(df['intent_to_admit_dttm'], utc=True)
ser = ser.dt.tz_convert('US/Pacific')
df['intent_to_admit_dttm'] = ser
df['intent_to_admit_dttm']
```

```python
ser = pd.to_datetime(df['inpatient_order_dttm'], utc=True)
ser = ser.dt.tz_convert('US/Pacific')
df['inpatient_order_dttm'] = ser
df['inpatient_order_dttm']
```

```python
admit_df = pd.DataFrame(index=pd.to_datetime(df['estimated_admission_dttm']))
admit_df['count_1'] = 1
admit_df = admit_df.resample('1d').sum()
admit_df
```

```python
admit_df.index = pd.to_datetime(admit_df.index)
admit_df['dayofweek'] = admit_df.index.dayofweek
admit_df
```

```python
admit_df = admit_df.groupby(['dayofweek']).mean()
admit_df
```

```python
admit_df.reset_index(inplace=True)
admit_df
```

```python
import seaborn as sns
sns.set_theme(style="whitegrid")
plotted = sns.catplot(data=admit_df, x="dayofweek", y="count_1", kind="bar", errorbar="sd", height=10)
plotted.set(ylim=(0, 10))
plotted.despine(left=True)
plotted.set_axis_labels('0 = Monday, 1 = Tuesday, 2 = Wednesday, 3 = Thursday, 4 = Friday, 5 = Saturday, 6 = Sunday', "Average # Admissions")
```
